#### implement a simple Bayesian Network model using the Graph Library, Networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
model = nx.DiGraph()
model.add_edges_from([('N', 'H'), ('W', 'H'), ('H', 'P'), ('W', 'O')])
nx.draw_networkx(model, edge_color='white')
plt.show()
```

#### to check for D-Separation in the above graph model

```python
nx.is_d_separator(model, {"H"}, {"O"}, {"W"})
```

```python
nx.is_d_separator(model, {"W"}, {"P"}, {"H"})
```

```python
nx.is_d_separator(model, {"N"}, {"W"}, {"H"})
```

```python
nx.is_d_separator(model, {"N"}, {"O"}, {"W"})
```
